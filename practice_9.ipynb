Чтобы начать думать о решении, возьмем для примера следующую которкую строку с оценками, записанными подряд:

Для решения задачи идем по строке, берем по порядку каждый символ, и если он подходит (удовлетворяет нужному условию, то есть является 4 или 5) - запоминаем, что встретили такой символ (увеличиваем счетчик).

##### timestamp: 1:08

Для начала напишем программу, которая считает, сколько встретилось четверок:

```python
a = "415254"

i = 0 # итератор для цикла
counter = 0 # переменная-счеткик, в которой будет храниться количество встреченных по ходу строки четверок

while i < 6: # во взятой для примера строке шесть символов, выполняем цикл для каждого
    if a[i] == '4': # условие "i-й элемент строки - это '4'";  
        counter = counter + 1 # при условии, что текущий (i-й) элемент строки подходит, увеличиваем счетчик
    i = i + 1 # увеличиваем значение итератора, чтобы цикл перешел к следующему элементу, а не выполнялся бесконечно
    
print(counter) # выводим значение счетчика, чтобы проверить, что все посчиталось правильно; во взятой для примера строке должно быть найдено две четверки
```

##### timestamp: 2:26

Как модифицировать программу, чтобы считать не только встреченные четверки, но и пятерки?

Для этого можно использовать оператор **or**. Условие "текущий элемент строки равен '5' или равен '5'" будет выглядеть так:

```python
# a[i] == '5' or a[i] == '4' # само по себе условие не нужно выполнять как программу, поэтому строка закоментирована (символ # делает все, что следует за ним, комментарием, а не исполняемым кодом)
```

Теперь значение счетчика должно быть равно 4, потому что во взятой для примера строки суммарно 4 положительные оценки (4 или 5):

```python
a = "415254"

i = 0 # итератор для цикла
counter = 0 # переменная-счеткик, в которой будет храниться количество встреченных по ходу строки четверок или пятерок

while i < 6: # во взятой для примера строке шесть символов, выполняем цикл для каждого
    if a[i] == '5' or a[i] == '4': # условие "i-й элемент строки - это '4'";
        counter = counter + 1 # при условии, что текущий (i-й) элемент строки подходит, увеличиваем счетчик
    i = i + 1 # увеличиваем значение итератора, чтобы цикл перешел к следующему элементу, а не выполнялся бесконечно
    
print(counter)
```

##### timestamp: 3:06

Для того, чтобы программа работала с любой введенной строкой, нужно:
1. заменить жестко заданную в коде строку на пользовальский ввод;
2. заменить жестко заданную в коде длину строки на посчитанную длину строки, введенной пользователем.

```python
a = input() # пользователь вводит любую строку с оценками
n = len(a) # длина строки, введенной пользователем
i = 0
counter = 0
while i < n: # цикл выполняется для каждого элемента строки, всего n раз (n - длина введенной строки)
    if a[i] == '5' or a[i] == '4':
        counter = counter + 1
    i = i + 1
print(counter)
```
