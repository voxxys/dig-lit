##### timestamp: 1:22

Чтобы выделить первое число, нужно считывать символы от начала строки и до тех пор, пока не встретится символ "+" - аналогично тому, как реализовано нахождение длины числовой строки в Практике 11.

Полученную подстроку нужно записать в переменную a, в которой будет храниться первое число для сложения.

После знака + все символы нужно записывать в другую переменную b.

##### timestamp: 2:02

Это можно реализовать, написав программу, которая может переключиться между двумя режимами: запись символов в переменную a и запись символов в переменную b.
    
Для переключения между этими двумя состояниями можно воспользоваться переменной состояния (флагом):

```python
a = "" # строка, в которую будем записывать символы первого числа
b = "" # строка, в которую будем записывать символы второго числа

com = "1234+34567" # пример введенной строки

flag = 0 # для переключения программы между состоянием записи в первое число и состоянием записи во второе число
# при flag == 0 будем записывать в первое число
# с начала строки начинаем записывать в первое число, поэтому изначально ставим flag = 0

i = 0
while i < len(com): # пока не дошли до конца строки (пока итератор не равен длине строки):
    if flag == 0: # в начале программы флаг 0 - записываем символы в строку для первого числа
        if com[i] != '+': # если еще не дошли до символа "+"
            a = a + com[i] # записываем символы в строку для первого числа
        else: # если дошли до "+"
            flag = 1 # переходим в состояние, где записываем во второе число
            # и больше ничего не делаем - символ "+" не нужно записывать ни в первое, ни во второе число
    else: # если флаг не равен 0, то есть мы переключились в запись второго числа
        b = b + com[i] # записываем символы в строку для второго числа
    i = i + 1 # увеличиваем значение итератора, чтобы цикл перешел к следующему символу

print(int(a)+int(b)) # переводим найденные строки в числа, складываем и выводим
            
        
```
