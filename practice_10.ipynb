##### timestamp: 0:22

Чтобы начать решать задачу, попробуем разбить ее на части. 

Например, представим, что у нас уже есть посчитанное среднее значение цифр в строке и теперь нужно посчитать количество цифр, превышающих это известное значение.

```python
# Подзадача: посчитать количество цифр в строке, превышающих некоторое значение
str1 = "123456" # взятая для примера строка
mean = 3.5 # известное для нее среднее значение
```

Как и прошлых задачах, создаем переменную, в которой будет храниться число встреченных символов, подходящих под условие, и итератор для цикла

```python
counter = 0 # создаем счетчик цифр, превышающих среднее
i = 0 # создаем итератор для цикла
```

Максимальное значение итератора (количество раз, которые должен выполниться цикл) можно подставить прямо в условие после while:

```python
while i < len(str1): # цикл выполняется для каждого символа строки (всего len(str1) раз - по количеству символов в строке)
    if int(str1[i]) > mean: # поскольку mean - число, а str[i] - символ, взятый из строки, нужно привести ее к числовому виду и взять int(str1[i])
        counter = counter + 1 # если условие выполняется, увеличиваем счетчик
    i = i + 1 # увеличиваем значение итератора, чтобы цикл перешел к следующему элементу, а не выполнялся бесконечно
print(counter) # на взятой для примера строке должно получиться 3
```

##### timestamp: 2:06

Среднее посчитано, теперь можно начать думать, как решить полную задачу.

Логично добавить часть, в которой вместо заданного в коде значения среднее будет вычисляться. Это задача, для которой нужно написать еще один цикл.

Этот цикл нужно будет добавить в начале программы, потому что результат (вычисленное среднее) будет использоваться в той части, которую мы написали выше.

Для того, чтобы найти среднее, нужно пройти по всем элементам строки и найти их сумму, а затем разделить на количество элементов (длину строки):

```python
mean = 0
i = 0

while i < len(str1): # цикл выполняется для каждого символа строки (всего len(str1) раз - по количеству символов в строке)
    mean = mean + int(str1[i]) # поскольку mean - число, а str[i] - символ, взятый из строки, нужно привести ее к числовому виду и взять int(str1[i])
    i = i + 1 # увеличиваем значение итератора, чтобы цикл перешел к следующему элементу, а не выполнялся бесконечно
    
print(mean)
```

Сейчас в переменной mean находится сумма всех элементов массива. Чтобы найти среднее, делим полученную сумму на количество чисел в строке:

```python
mean =  mean/len(str1)
print(mean)
```

Объединим обе части программы, чтобы получить полное решение задачи:

```python
str1 = "123456" # взятая для примера строка, ее нужно задать в самом начале программы, поскольку ее используют оба цикла

mean = 0 # переменная, в которой будет храниться сначала сумма всех пройденных элементов строки, а потом (после прохождения цикла) среднее
i = 0 # создаем итератор для цикла

while i < len(str1): # цикл выполняется для каждого символа строки (всего len(str1) раз - по количеству символов в строке)
    mean = mean + int(str1[i]) # поскольку mean - число, а str[i] - символ, взятый из строки, нужно привести ее к числовому виду и взять int(str1[i])
    i = i + 1 # увеличиваем значение итератора, чтобы цикл перешел к следующему элементу, а не выполнялся бесконечно

mean =  mean/len(str1) # чтобы найти среднее, делим полученную сумму на количество чисел в строке

counter = 0 # создаем счетчик циф, превышающих среднее

while i < len(str1): # цикл выполняется для каждого символа строки (всего len(str1) раз - по количеству символов в строке)
    if int(str1[i]) > mean: # поскольку mean - число, а str[i] - символ, взятый из строки, нужно привести ее к числовому виду и взять int(str1[i])
        counter = counter + 1 # если условие выполняется, увеличиваем счетчик
    i = i + 1 # увеличиваем значение итератора, чтобы цикл перешел к следующему элементу, а не выполнялся бесконечно
print(counter) # на взятой для примера строке должно получиться 3
```

Попробуйте запустить получившийся код. Выдает ли он ожидаемый результат?

Если по шагам проследить значение переменной i на протяжении исполнения программы, можно увидеть, что она используется в обоих циклах и при этом не обнуляется перед вторым циклом.

По этой причине второй цикл начинает проходить строку не с первого элемента, 
а с элемента с номером, как раз вышедшим за длину строки (при выполнении прошлого цикла) - условие while сразу не выполняется и цикл не запускается ни разу.

Чтобы программа заработала, нужно обнулить значение итератора перед вторым циклом:

```python
str1 = "123456" # взятая для примера строка, ее нужно задать в самом начале программы, поскольку ее используют оба цикла

mean = 0 # переменная, в которой будет храниться сначала сумма всех пройденных элементов строки, а потом (после прохождения цикла) среднее
i = 0 # создаем итератор для цикла

while i < len(str1): # цикл выполняется для каждого символа строки (всего len(str1) раз - по количеству символов в строке)
    mean = mean + int(str1[i]) # поскольку mean - число, а str[i] - символ, взятый из строки, нужно привести ее к числовому виду и взять int(str1[i])
    i = i + 1 # увеличиваем значение итератора, чтобы цикл перешел к следующему элементу, а не выполнялся бесконечно

mean =  mean/len(str1) # чтобы найти среднее, делим полученную сумму на количество чисел в строке

counter = 0 # создаем счетчик циф, превышающих среднее

i = 0 # обнуляем итератор!

while i < len(str1): # цикл выполняется для каждого символа строки (всего len(str1) раз - по количеству символов в строке)
    if int(str1[i]) > mean: # поскольку mean - число, а str[i] - символ, взятый из строки, нужно привести ее к числовому виду и взять int(str1[i])
        counter = counter + 1 # если условие выполняется, увеличиваем счетчик
    i = i + 1 # увеличиваем значение итератора, чтобы цикл перешел к следующему элементу, а не выполнялся бесконечно
print(counter) # на взятой для примера строке должно получиться 3
```
